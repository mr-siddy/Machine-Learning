# Subplots

```python
plt.subplot(2,2,1) # 2 rows 2 cols 1 position
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
np.pi
```

```python
x = np.arange(0,4*np.pi,0.1)
y=np.sin(x)

plt.title("sine wave form")

plt.plot(x,y)
plt.show()
```

```python
#subplot for sin and cos waves

x=np.arange(0,5*np.pi,0.1)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title("sine graph")

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title("cosine graph")

plt.show()
```

# Bar plot

```python
x= [2,8,10]
y = [11,16,18]

x2 = [3,9,11]
y2 = [4,7,9]

plt.bar(x,y)
plt.bar(x2,y2,color ='g')

plt.title('Bar graph')
plt.ylabel( 'Yaxis')
plt.xlabel( 'Xaxis')

plt.show()
```

# Histograms

```python
 a = np.array([1,2,3,4,5,5,6,67,7,8,8,9])  # y axis == bins - desity or count

plt.hist(a)
plt.title('histogram')
plt.show()
```

# Box plot

```python
# helps to find percentiles

data = [np.random.normal(0,std,100) for std in range(1,4)] # selecting a normal distribution b/w low=0, to std, step=100

# rectangular box plot
plt.boxplot(data, vert=True, patch_artist= True)
```

# Pi chart

```python
labels = 'python','c++', 'ruby', 'java'
sizes = [215,130,245,210]
colors = ['gold', 'yellowgreen','lightcoral', 'lightskyblue']
explode = (0.1,0,0,0) #explode 1st slice

#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors, autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
