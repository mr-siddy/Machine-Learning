# Simple, Ridge and Lasso Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
```

```python
df = load_boston()
```

```python
df
```

```python
dataset= pd.DataFrame(df.data)
print(dataset.head())
```

```python
 dataset.head()
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
df.target.shape
```

```python
dataset["price"]=df.target
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,:-1] # independent fratures

y=dataset.iloc[:,-1] # dependent frature
```

# Linear Regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lin_regressor = LinearRegression()
mse = cross_val_score(lin_regressor,X,y,scoring='neg_mean_squared_error',cv=5) #cv=crossvalidation
mean_mse = np.mean(mse)
print(mean_mse) # this value should be nearer to zero
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}

ridge_regressor=GridSearchCV(ridge,parameters,scoring = 'neg_mean_squared_error',cv=5)

ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso=Lasso()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}

Lasso_regressor=GridSearchCV(lasso,parameters,scoring = 'neg_mean_squared_error',cv=5)

Lasso_regressor.fit(X,y)

print(Lasso_regressor.best_params_)
print(Lasso_regressor.best_score_)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
prediction_lasso=Lasso_regressor.predict(X_test)
prediction_ridge=ridge_regressor.predict(X_test)
```
