# Inbuilt functions

```python
# arange

arr = np.arange(0,10)
```

```python
arr
```

```python
arr = np.arange(0,10,step=2)
arr
```

```python
# linspace
# shift+tab for elaborate the function


np.linspace(1,10,50)
```

```python
# Copy function and broadcasting

print(arr)

arr[3:] =100 # replace all indexes starting from 3rd to all by 100

print(arr)
```

```python
arr1=arr
```

```python
arr1[3:]=500
arr1
```

```python
arr # array is actually a reference type hence change is reflected to actual array also
```

```python
# in order to prevent updation in original array also we use copy function

arr1 = arr.copy()
arr1
```

```python
arr1[3:] = 800
print(arr1)
print(arr)
```
