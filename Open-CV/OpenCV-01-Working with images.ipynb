```python
import cv2
import matplotlib.pyplot as plt
```

```python
# read image 

img = cv2.imread('dog.png') # dog.png have 3 channels RGB but when opencv reads images by default it reads as BGR 
plt.imshow(img) # so here red color of dog.png is swapped by blue
plt.show()
```

```python
# To switch b/w color spaces

newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(newImg)
plt.show()
```
