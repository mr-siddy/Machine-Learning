```python
import seaborn as sns 
import numpy as np 
import pandas as pd 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Distribution plot

sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips["total_bill"],kde = False,bins= 40)
```

```python
# kde plot()

sns.kdeplot(tips["total_bill"])
```

```python
# joint plot()

sns.jointplot(x= 'total_bill', y= 'tip', data = tips) #kind = default = scatter

# we can observe a linear trend b/w tip and total bill
```

```python
# to change the kind of graph 

sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
 # kind = 'kde'
    
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
```

```python
# pairplot() - compares all features from a dataframe

sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue = 'sex')
```
